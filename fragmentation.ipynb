```python
import pandas as pd
import numpy as np

summary6 = open("summary6.csv", "r")

rem_text = summary6.read()
text_split = rem_text.split('\n')
rem_data = np.array(text_split[0].split(','))

for i in range(1, len(text_split)-1):
    line = text_split[i].split(',')
    rem_data = np.vstack((rem_data, line))
        
pd.set_option('display.max_columns', None)
df = pd.DataFrame(rem_data[1:], columns=rem_data[0])
df
```

```python
# AHI 與 全部arousal fragmentation的分布圖

import matplotlib.pyplot as plt
%matplotlib inline

o_ahi = np.array(df['o_ahi'], dtype='single')
o_a1 = np.array(df['o_a1'], dtype='int64')
o_a2 = np.array(df['o_a2'], dtype='int64')
o_a3 = np.array(df['o_a3'], dtype='int64')
o_a4 = np.array(df['o_a4'], dtype='int64')
o_rem_duration = np.array(df['o_rem_duration'], dtype='int64')

x = o_ahi
y = ((o_a1 + o_a2 + o_a3 + o_a4) / (o_rem_duration/30)) * 100

plt.figure(figsize=(10,10))
plt.rcParams['font.sans-serif'] = ['Taipei Sans TC Beta']
plt.scatter(x, y, c='orange')
plt.axvline(x=15, c='red', linewidth=1)
plt.axvline(x=30, c='red', linewidth=1)

plt.xticks(fontsize=20)
plt.yticks(np.arange(0, 60, 5), fontsize=20)
plt.ylabel("fragmentation(%)", fontsize=20)
plt.xlabel("AHI", fontsize=20)
plt.title("暴露噪音下", fontsize=20)
plt.show()

x_ahi = np.array(df['x_ahi'], dtype='single')
x_a1 = np.array(df['x_a1'], dtype='int64')
x_a2 = np.array(df['x_a2'], dtype='int64')
x_a3 = np.array(df['x_a3'], dtype='int64')
x_a4 = np.array(df['x_a4'], dtype='int64')
x_rem_duration = np.array(df['x_rem_duration'], dtype='int64')

x = x_ahi
y = ((x_a1 + x_a2 + x_a3 + x_a4) / (x_rem_duration/30)) * 100

plt.figure(figsize=(10,10))
plt.rcParams['font.sans-serif'] = ['Taipei Sans TC Beta']
plt.scatter(x, y, c='blue')
plt.axvline(x=15, c='red', linewidth=1)
plt.axvline(x=30, c='red', linewidth=1)

plt.xticks(fontsize=20)
plt.yticks(np.arange(0, 60, 5), fontsize=20)
plt.ylabel("fragmentation(%)", fontsize=20)
plt.xlabel("AHI", fontsize=20)
plt.title("非暴露噪音下", fontsize=20)
plt.show()
```
