```python
from tkinter import *
root = Tk()
w, h = root.maxsize()
root.geometry("{}x{}".format(w, h))

dirName = "P006_x_20141206_1"

# 讀stage檔
stageFile = open("G:\共用雲端硬碟\Sleep center data\REM片斷化\\"+dirName+"\hyp_data.txt")
stage = stageFile.read()
eventFile = open("G:\共用雲端硬碟\Sleep center data\REM片斷化\\"+dirName+"\event_data.txt")
event = eventFile.read()

stage = stage.split('\n')[:-1]
#print(stage)
epoch = len(stage)
print(epoch)

event = event.split('\n')[:-1]
event_epoch = []
event_name = []
event_stage = []
#print(event)
for i in range(len(event)):
    event_epoch.append(event[i].split(',')[1])
    event_stage.append(event[i].split(',')[2])
    event_name.append(event[i].split(',')[3])
    
arousal_epoch = []
arousal_name = []
arousal_stage = []
for i in range(len(event_name)):
    if event_name[i] == "Arousal 3 ARO SPONT":
        arousal_epoch.append(event_epoch[i])
        arousal_name.append(event_name[i])
        arousal_stage.append(event_stage[i])

cv = Canvas(root,bg = 'white', width=1800, height=1000)
cv.pack()

# 畫總長度
cv.create_line(50, 200, epoch+50, 200)
cv.create_line(50, 200, 50, 80)
cv.create_line(50, 80, 60, 80)
cv.create_text(35, 80, text="R")
cv.create_line(50, 110, epoch+50, 110)
cv.create_text(35, 110, text="W")
cv.create_line(50, 140, 60, 140)
cv.create_text(35, 140, text="N1")
cv.create_line(50, 170, 60, 170)
cv.create_text(35, 170, text="N2")
cv.create_text(35, 200, text="N3")
# 畫刻度
for i in range(int(epoch/100)):
    cv.create_line(150+100*i, 200, 150+100*i, 190)
    cv.create_text(150+100*i, 210, text=str(100+100*i))
# 畫階段
for i in range(epoch):
    if stage[i] == 'W':
        pass
    elif stage[i] == '1' or stage[i] == 'N1':
        cv.create_rectangle(i+50, 110, i+51, 140, fill='#edb120', outline='#edb120')
    elif stage[i] == '2' or stage[i] == 'N2':
        cv.create_rectangle(i+50, 110, i+51, 170, fill='#77ac30', outline='#77ac30')
    elif stage[i] == '3' or stage[i] == 'N3':
        cv.create_rectangle(i+50, 110, i+51, 200, fill='#0072bd', outline='#0072bd')
    elif stage[i] == 'R':
        cv.create_rectangle(i+50, 110, i+51, 80, fill='#a2142f', outline='#a2142f')
# 畫arousal
print(arousal_epoch)
print(arousal_name)
print(arousal_stage)
for i in range(len(arousal_name)):
    cv.create_line(int(arousal_epoch[i])+50, 90, int(arousal_epoch[i])+50, 190, fill='gray', width=1)


root.mainloop()
```
